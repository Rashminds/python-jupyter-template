### Exercise 4: Pie Chart with Percentages

```python

import matplotlib.pyplot as plt


labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]


plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)


plt.axis('equal')


plt.title('Programming Languages Pie Chart')


plt.show()
```

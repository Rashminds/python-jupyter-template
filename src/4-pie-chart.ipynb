### Exercise 4: Pie Chart with Percentages

```python
try:
    import matplotlib.pyplot as plt
except ImportError:
    print("Error: Matplotlib is not installed. Install it using 'pip install matplotlib'")
    exit()

def validate_numeric_data(data, data_name="Data"):
    if not isinstance(data, list):
        raise ValueError(f"{data_name} should be a list")
    if not all(isinstance(i, (int, float)) for i in data):
        raise ValueError(f"All elements in {data_name} must be numbers")

try:
    labels = ['Python', 'Java', 'C++', 'JavaScript']
    sizes = [40, 25, 20, 15]

    
    if not all(isinstance(label, str) for label in labels):
        raise ValueError("All labels must be strings")
    validate_numeric_data(sizes, "sizes")

    
    if sum(sizes) != 100:
        print("Warning: Sizes do not add up to 100")

    plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
    plt.axis('equal')
    plt.title('Pie Chart with Percentages')
    plt.show()
except Exception as e:
    print(f"An error occurred in pie chart: {e}")
```

```python
try:
    import matplotlib.pyplot as plt
except ImportError:
    print("Error: Matplotlib is not installed. Install it using 'pip install matplotlib'")
    exit()

def validate_numeric_data(data, data_name="Data"):
    if not isinstance(data, list):
        raise ValueError(f"{data_name} should be a list")
    if not all(isinstance(i, (int, float)) for i in data):
        raise ValueError(f"All elements in {data_name} must be numbers")

try:
   
    x = [1, 2, 3, 4, 5]
    y = [i**2 for i in x]

    validate_numeric_data(x, "x")
    validate_numeric_data(y, "y")

    plt.plot(x, y, marker='o')
    plt.title('Line Plot of y = x^2')
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.show()
except Exception as e:
    print(f"An error occurred in line plot: {e}")
```
